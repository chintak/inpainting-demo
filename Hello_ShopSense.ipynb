- Structural inpainting using Telea's (2004) Fast Marching Method.

Inpainting using Exemplar Based Method
--------------------------------------

In this example we wll show linear structural and textural inpainting. There
are two terms which play a role in determining the order of inpainting:

- ``confidence`` signifies the amount of reliable information in the
  neighbourhood. Already known pixels have ``confidence = 1`` and unknown
  pixels ``0``.
- ``data`` term represents the presence of strong edges hitting the mask
  boundary. This implementation updates 1 patch at a time.


```python
from skimage.filter import inpaint_criminisi
# inpaint_criminisi?
from skimage.data import checkerboard
from skimage.util import img_as_ubyte

# Generate a synthetic image and mask
image = img_as_ubyte(checkerboard())
mask = np.zeros_like(image, dtype=np.uint8)
paint_region = (slice(50, 150), slice(50, 150))
image[paint_region] = 0
mask[paint_region] = 1

# Display the input image and the mask
fig, (ax0, ax1) = plt.subplots(ncols=2)
ax0.set_title('Input image')
ax0.imshow(image, cmap=plt.cm.gray)
ax1.set_title('Binary Mask')
ax1.imshow(mask, cmap=plt.cm.gray)
```

```python
painted = inpaint_criminisi(image, mask, window=27, ssd_thresh=0.2)
plt.imshow(painted, cmap=plt.cm.gray)
```

Inpainting using Fast Marching Method
-------------------------------------

In this example we wll show Structural inpainting which uses geometric
approaches for filling in the missing information in the region which should
be inpainted. These algorithms focus on the consistency of the geometric
structure.

```python
from skimage import data
from skimage.filter import inpaint

image = data.camera()
paint_region = (slice(240, 260), slice(360, 420))
paint_region1 = (slice(430, 450), slice(360, 410))

mask = np.zeros_like(image, dtype=np.uint8)
mask[paint_region] = 1
mask[paint_region1] = 1
image[mask == 1] = 0

fig, (ax0, ax1) = plt.subplots(ncols=2)
ax0.set_title("Input Image")
ax0.imshow(image, cmap=plt.cm.gray)
ax1.set_title("Inpainted Image")
ax1.imshow(mask, cmap=plt.cm.gray)
plt.show()
```

```python
painted = inpaint.inpaint_fmm(image, mask)
plt.imshow(painted, cmap=plt.cm.gray)
```
